```python
import pandas as pd

df = pd.read_csv('responses.csv')
```

```python
from textblob import TextBlob

def get_sentiment(text):
    if TextBlob(str(text)).sentiment.polarity > 0:
        return "Positive"
    elif TextBlob(str(text)).sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df['Sentiment'] = df['What did you like about the event?'].apply(get_sentiment)

df[['What did you like about the event?', 'Sentiment']].head()
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df['Sentiment'].value_counts()

sentiment_counts.plot(kind='bar')

plt.title('Sentiment Analysis of Event Feedback')
plt.xlabel('Sentiment')
plt.ylabel('Number of Responses')

plt.show()
```
